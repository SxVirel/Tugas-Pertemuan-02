```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.info()
```

```python
kolomdihapus = ['Unnamed: 0', 'ISRC', 'TIDAL Popularity']
```

```python
df.drop(columns=kolomdihapus, inplace=True)
```

```python
df.info()
```

```python
numeric_columns = [
    'All Time Rank','Track Score', 'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach',
    'Spotify Popularity', 'YouTube Views', 'YouTube Likes', 'TikTok Posts', 'TikTok Likes',
    'TikTok Views', 'YouTube Playlist Reach', 'Apple Music Playlist Count', 'AirPlay Spins',
    'SiriusXM Spins', 'Deezer Playlist Count', 'Deezer Playlist Reach', 'Amazon Playlist Count',
    'Pandora Streams', 'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts'
]
```

```python
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df.rename(columns={'Release Date': 'Year Released'}, inplace=True)
```

```python
df['Year Released'] = pd.to_datetime(df['Year Released'], format='%m/%d/%Y').dt.year
```

```python
df
```

```python
df.info()
```

```python
df['Track'].unique()
```

```python
df['Year Released'].value_counts().plot(kind='bar')
```

```python
df_encoded = pd.get_dummies(df, columns=['Artist', 'Album Name'], drop_first=True)
df_numerik = df_encoded.select_dtypes(include=['float64', 'int32', 'int64'])
plt.figure(figsize=(12,10))
sns.heatmap(df_numerik.corr(), annot=True, annot_kws={'size': 15}, cmap='coolwarm')
plt.title('Heatmap Korelasi dengan Kolom Objek')
plt.show()
```

```python
plt.figure(figsize=(50, 20))
sns.boxplot(data=df[numeric_columns])
plt.show()
```

```python
sns.boxplot(data=df['Track Score'])
plt.title('Track Score Boxplot')
plt.show()
```

```python
df['Track Score'] = np.log(df['Track Score'])

sns.boxplot(df['Track Score'])
plt.show()

sns.histplot(df['Track Score'], bins=35)
plt.show()
```
